## CHURN RATE BY VOICE MAIL PLAN

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate churn rate by voice mail plan
churn_stats = df.groupby('voice_mail_plan').agg({
    'churn': ['sum', 'count', 'mean']
}).reset_index()
churn_stats.columns = ['voice_mail_plan', 'churned', 'total', 'churn_rate']
churn_stats['churn_rate_pct'] = churn_stats['churn_rate'] * 100

# Create figure
plt.figure(figsize=(10, 6))

# Colors: Red for higher churn, Green for lower churn
colors = ['#e74c3c', '#27ae60']  # No plan = higher churn (red), Has plan = lower (green)

bars = plt.bar(['No Voice Mail Plan', 'Has Voice Mail Plan'], 
               churn_stats['churn_rate_pct'], 
               color=colors, edgecolor='black', linewidth=2, width=0.5)

# Add percentage labels on bars
for bar, rate, count, churned in zip(bars, churn_stats['churn_rate_pct'], 
                                       churn_stats['total'], churn_stats['churned']):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.8, 
             f'{rate:.1f}%', ha='center', fontsize=16, fontweight='bold')
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, 
             f'{churned}/{count}\ncustomers', ha='center', fontsize=11, color='white', fontweight='bold')

# Add horizontal line for average churn rate
plt.axhline(y=14.5, color='blue', linestyle='--', linewidth=2, label='Overall Avg (14.5%)')

plt.xlabel('Voice Mail Plan Status', fontsize=12, fontweight='bold')
plt.ylabel('Churn Rate (%)', fontsize=12, fontweight='bold')
plt.title('Churn Rate by Voice Mail Plan', fontsize=14, fontweight='bold')
plt.ylim(0, 25)
plt.legend(loc='upper right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Inference
# Customers WITHOUT voice mail plan: ~16-17% churn rate (above average)
# Customers WITH voice mail plan: ~8-9% churn rate (below average)
# Voice mail plan reduces churn by nearly 50%!
# Voice mail acts as a "sticky feature" - increases engagement and retention
# Recommendation: Offer free voice mail trial to at-risk customers
```

## CHURN RATE BY CUSTOMER SERVICE CALL

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate churn rate by service calls
churn_stats = df.groupby('customer_service_calls').agg({
    'churn': ['sum', 'count', 'mean']
}).reset_index()
churn_stats.columns = ['service_calls', 'churned', 'total', 'churn_rate']
churn_stats['churn_rate_pct'] = churn_stats['churn_rate'] * 100

# Create figure
plt.figure(figsize=(12, 7))

# Color bars based on risk level
colors = ['#27ae60' if x < 25 else '#f39c12' if x < 40 else '#e74c3c' 
          for x in churn_stats['churn_rate_pct']]

bars = plt.bar(churn_stats['service_calls'], churn_stats['churn_rate_pct'], 
               color=colors, edgecolor='black', linewidth=1.2)

# Add percentage labels on bars
for bar, rate in zip(bars, churn_stats['churn_rate_pct']):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1.5, 
             f'{rate:.0f}%', ha='center', fontsize=11, fontweight='bold')

# Add horizontal line for average churn rate
plt.axhline(y=14.5, color='blue', linestyle='--', linewidth=2, label='Overall Avg (14.5%)')

# Add danger zone shading
plt.axhspan(40, 100, alpha=0.1, color='red', label='High Risk Zone (>40%)')

plt.xlabel('Number of Customer Service Calls', fontsize=12, fontweight='bold')
plt.ylabel('Churn Rate (%)', fontsize=12, fontweight='bold')
plt.title('Churn Rate by Number of Customer Service Calls', fontsize=14, fontweight='bold')
plt.xticks(range(0, 10))
plt.ylim(0, 100)
plt.legend(loc='upper left')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Inference
# Customers with 0-2 calls: Low churn (10-15%) - Satisfied customers
# Customers with 3 calls: Medium risk (20-25%) - Growing frustration
# Customers with 4+ calls: High risk (40-80%+) - Immediate intervention needed
# Recommendation: Flag customers after 2nd call for proactive support
```

```python
# Boxplots for quick outlier visualization
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,8))
sns.boxplot(data=df, orient="h", palette="Set2")
plt.title("Outlier Detection (Boxplots)")
plt.show()
# sns.boxplot  draws boxplots for all numeric columns in your DataFrame (df).
# data=df  uses your whole dataset.
# orient="h"  horizontal boxplots (easier to read when you have many features).
# palette="Set2"  color scheme for styling.

# creates horizontal boxplots for all numeric features in your dataset
# so you can visually spot outliers, skewness, and spread

# We retained outliers because tree-based models like XGBoost are inherently robust to 
# them. More importantly, in churn prediction, outliers often represent at-risk customers
# removing them would mean losing critical predictive signals.
```

```python
# Simple Outlier Detection per Column (IQR Method)
import numpy as np
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

for col in df.select_dtypes(include=[np.number]).columns:
    lower = Q1[col] - 1.5 * IQR[col]
    upper = Q3[col] + 1.5 * IQR[col]
    
    outliers = df[(df[col] < lower) | (df[col] > upper)][col]
    
    print(f"\nColumn: {col}")
    print(f"Number of Outliers: {outliers.shape[0]}")
    print("Outlier Values:")
    print(outliers.values[:10])  # show first 10 outlier values


# For each numeric column:
# Q1 = 25th percentile (lower quartile).
# Q3 = 75th percentile (upper quartile).
# IQR = Q3  Q1  Interquartile Range (spread of middle 50% of the data).

# for col in df.select_dtypes(include=[np.number]).columns:
# Loops over all numeric columns only (ignores categorical/text).

# lower = Q1[col] - 1.5 * IQR[col]
# upper = Q3[col] + 1.5 * IQR[col]
# Defines the outlier cutoff range:
# Anything below Q1  1.5  IQR is a low outlier.
# Anything above Q3 + 1.5  IQR is a high outlier.

# outliers = df[(df[col] < lower) | (df[col] > upper)][col]
# Filters the dataset and extracts all outlier values for that column.

# in short: 
# applies the IQR method to find values that are unusually high/low, 
# then prints how many outliers exist and shows a preview of them

```

## SMOTE ANALYSIS

```python
# Import library
from imblearn.over_sampling import SMOTE
from collections import Counter

# Before SMOTE
print("Before SMOTE:", Counter(y))

# Apply SMOTE
sm = SMOTE(random_state=42)
x_resampled, y_resampled = sm.fit_resample(x_scaled, y)

# After SMOTE
print("After SMOTE:", Counter(y_resampled))

#Before SMOTE: {0: 2850, 1: 483}
#0 = customers who did not churn  2850
#1 = customers who churned  483
#This is imbalanced (many more non-churners than churners).

#After SMOTE: {0: 2850, 1: 2850}
#Now, churners (1) were synthetically increased to match non-churners (0).
#Instead of just duplicating rows, SMOTE creates new synthetic samples of churners by 
#interpolating between existing churn data points.

#Why do we do this?
#Balance the dataset  so the model doesnt get biased toward predicting No churn.
# Improve accuracy for minority class  the model will learn patterns of churners better.
# Fair training  both churn and non-churn customers get equal importance.
```

## selecting best features using RFC

```python
#using random forest classifier algorithm we are selceting top 15 features which has most impact on the churn 
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

rf = RandomForestClassifier(random_state=42)
rf.fit(x, y)

# Get importance scores
importances = pd.Series(rf.feature_importances_, index=x.columns)

# Sort & show top 10 features
best_features = importances.sort_values(ascending=False).head(10)
print(best_features)

#as a result we get this 10 features printed `
```

## Train test split

```python
# 1) Train-test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=28)

# here during the train_test_split
# 80% of the data (used to train the model).
#  20% of the data (used to test/evaluate the model).

#  Parameters explained
# test_size=0.2  means 20% of the dataset goes to testing, 80% to training.
# random_state=28  ensures the split is the same every time you run it (reproducibility).
# Without random_state, youd get different train/test splits each time.

# Why we do this
# The model learns patterns from training data.
# Then we test it on unseen test data to check performance.
# This prevents overfitting (model memorizing instead of learning).

```

## model buiding

```python
#here in the model building we juz applied RFC algorithm
#RandomForestClassifier is an ensemble learning method: it builds many decision trees 
#and combines their predictions for better accuracy and stability.

#why random forest 
#it avoids overfitting
#it handles imbalanced data
#it selects the best features 
#handles non liner relation
#provides high accuracy
```

```python
# 2) Build RandomForestClassifier model
from sklearn.ensemble import RandomForestClassifier
# Initialize the Random Forest model
model = RandomForestClassifier(n_estimators=100,max_depth=None,random_state=42)    
# Fit the model
model.fit(x_train,y_train)

# This initializes a Random Forest model with three main parameters:
# n_estimators=100  The forest will have 100 decision trees. More trees usually improve performance .

# max_depth=None  Trees are grown fully until leaves are pure or until each leaf 
#has less than min_samples_split samples. Setting a limit can prevent overfitting.

# random_state=42  Ensures reproducibility. 


# This trains the Random Forest model using  training data:
# x_train  features (independent variables)
# y_train  target variable (labels you want to predict)
# During .fit(), the model builds 100 decision trees, each trained on a random subset of data and features.
# After this, model learns patterns from the data so it can predict unseen samples.

```

## prediction

```python
# Make predictions on unseen dataset (test dataset)
y_pred = model.predict(x_test)
y_pred
```

## evaluation

```python
#evaluation : 
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
#out of 667 test datset 20% 
```

```python
# ============================================
# 1. Confusion Matrix Heatmap
# ============================================
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No Churn (0)', 'Churn (1)'],
            yticklabels=['No Churn (0)', 'Churn (1)'])
plt.title('Confusion Matrix', fontsize=14, fontweight='bold')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
